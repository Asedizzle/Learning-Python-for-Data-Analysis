```python
import matplotlib.pyplot as plt, seaborn as sns
%matplotlib inline

work_hours_Abena = [4,6,4,7,9]
work_hours_Tracy = [3,6,8,7,10]
days = ["Mon", "Tue", "Wed", "Thur", "Fri"]
```

```python
plt.xlabel("Working hours")
plt.ylabel("Working days")
plt.plot(work_hours_Abena, days);
plt.plot(work_hours_Tracy, days);
plt.title("Working hours per day")
plt.legend(['Abena', "Tracy"])

```

```python
hours = [8,9,10,11,12]
num_red_cars = [9,7,5,4,1]
num_blue_cars = [2,5,7,9,11]
```

Styling Lines and Markers
The plt.plot function supports many arguments for styling lines and markers:

color or c: Set the color of the line (supported colors)
linestyle or ls: Choose between a solid or dashed line
linewidth or lw: Set the width of a line
markersize or ms: Set the size of markers
markeredgecolor or mec: Set the edge color for markers
markeredgewidth or mew: Set the edge width for markers
markerfacecolor or mfc: Set the fill color for markers
alpha: Opacity of the plot

plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)

```python
plt.figure(figsize=(12,8))
plt.plot(num_red_cars, hours, marker="o", c='r', ms=6, lw=3, ls="-", mfc="black")
plt.plot(num_blue_cars, hours, marker="s", c='b', ms=6, lw=3, ls="-", mfc="black")
plt.title("A graph showing the number of cars in a certain colour on a road")
plt.ylabel("Number of cars")
plt.xlabel("Time of day")
plt.legend(["red cars", "blue cars"])
```

```python
#The fmt argument provides a shorthand for specifying the marker shape, line style, and line color. It can be provided as the third argument to plt.plot.
plt.figure(figsize=(10,6))
plt.plot(num_red_cars, hours, "s--g")
plt.plot(num_blue_cars, hours, "o-y")

```

```python
#If you don't specify a line style in fmt, only markers are drawn
plt.plot(num_blue_cars, hours, "og")
```

```python
'''Changing the Figure Size
You can use the plt.figure function to change the size of the figure. Make sure to put plt.figure before the plot command'''

plt.figure(figsize=(12,10))
plt.plot(num_blue_cars, hours, 'o-y', ms= 4 )
plt.title("Number of blue cars per day")
plt.legend(["blue cars"])
```
