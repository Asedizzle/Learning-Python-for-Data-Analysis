```python
#The result of performing an operation on two columns is a new series.
average_deaths = round(cd_data['new_deaths'].mean(),2)
average_deaths
```

```python
#The result of performing an operation on two columns is a new series.
ratio_cases_totests = cd_data['new_cases']/cd_data['new_tests']
ratio_cases_totests 
```

```python
#To add as a new column
cd_data['ratio_cases_totests '] = cd_data['new_cases']/cd_data['new_tests']
cd_data
```

```python
#To remove columns use .drop()
cd_data.drop(columns=['ratio_cases_totests '], inplace=True)
cd_data
```
