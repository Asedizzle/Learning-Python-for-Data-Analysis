The data type of date is currently object, so Pandas does not know that this column is a date. We can convert it into a datetime column using the pd.to_datetime method.

```python

df['date'] = pd.to_datetime(df.date)
```

```python
df['date']
```

```python
df['year'] = pd.DatetimeIndex(df.date).year
df['month'] = pd.DatetimeIndex(df.date).month
df['day'] = pd.DatetimeIndex(df.date).day
df['weekday'] = pd.DatetimeIndex(df.date).weekday
```

```python
df
```

```python
# Query the rows for May
df_May = df[df.month == 5]
df_May
```

```python
# Extract the subset of columns to be aggregated
covid_df_may_metrics = df_May[['new_cases', 'new_deaths', 'new_tests']]
covid_df_may_metrics
```

```python
# Get the column-wise sum
covid_may_totals = covid_df_may_metrics.sum()
covid_may_totals
```

```python
#We can also combine the above operations into a single statement.
df[df.month == 5][['new_cases', 'new_deaths', 'new_tests']].sum()
```

```python
#Finding average for all days
day_average = df.new_cases.mean()
day_average
```

```python
#Finding average for sundays
saturday_average = df[df['weekday']== 6][['new_tests', 'new_cases', 'new_deaths']].mean()
saturday_average
```

```python
#Grouping
summary_day_data = df.groupby('month')[['new_cases', 'new_tests', 'new_deaths']].sum()
summary_day_data
```

```python
#IWe can also aggregate by mean. Try to  compute the average number of daily new cases, deaths, and tests for each month.
summary_databymean = df.groupby('month')[['new_tests','new_cases', 'new_deaths']].mean()
summary_databymean 
```
