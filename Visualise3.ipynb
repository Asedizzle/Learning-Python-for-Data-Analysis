```python
import matplotlib.pyplot as plt
import seaborn as sns 
flowers_db = sns.load_dataset("iris")
```

```python
flowers_db
```

```python
flowers_db.species.unique()
```

```python
plt.figure(figsize=(10,8))
plt.plot(flowers_db['sepal_length'], flowers_db['sepal_width']);
```

```python
sns.scatterplot(x=flowers_db['sepal_length'], y=flowers_db['sepal_width']);
```

```python
#Adding Hues
plt.figure(figsize=(12,8))
plt.title("A graph showing the sepal width and lenght of various flower species")
sns.scatterplot(x=flowers_db['sepal_length'],y=flowers_db['sepal_width'], hue=flowers_db['species'], s=100);
```

```python
#Seaborn has in-built support for Pandas data frames. Instead of passing each column as a series, you can provide column names and use the data argument to specify a data frame.
plt.figure(figsize = (12,5))
plt.title("Sepal dimensiosn")
sns.scatterplot(x= 'sepal_length' , y= 'sepal_width', hue = 'species', s = 100, data = flowers_db)
```
